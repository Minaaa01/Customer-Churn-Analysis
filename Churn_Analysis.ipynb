```python
def hist(df, group, target):
    """
    Function to generate a hist between two variables
    """
    churn_yes=df[df[target]=="Yes"]
    churn_no=df[df[target]=="No"]
    plt.hist([churn_yes[group],churn_no[group]],color=['red','green'],label=['Churn = Yes','Churn = No'])
    plt.xlabel(group)
    plt.ylabel("Number of customers")
    plt.legend()
    plt.title(f'{group}{" Figure"}')
    plt.xticks(rotation='vertical')
    plt.show()
```

```python
hist(df, "SeniorCitizen", "Churn")
hist(df, "Dependents", "Churn")
hist(df, "Partner", "Churn")
```

**From above plots, we can say**

*   The Non-Senior Citizens are high churners
*   If the customer has family, he or she is less likely to stay





```python
plt.figure(figsize=(16,8))
sns.countplot(x=df['tenure'],hue=df["Churn"])
plt.xticks(rotation='vertical')
plt.title("Fig 2")
plt.show()
```



*  This Figure tell us that the customer becomes loyal if his tenure was high, as we see that higher tenure, the churn rate become low.



```python
def tenure(t):
    if t<=12:
        return "Less Than One Year"
    elif t>12 and t<=24:
        return "2 Year"
    elif t>24 and t<=36:
        return "3 Years"
    elif t>36 and t<=48:
        return "4 Years"
    elif t>48 and t<=60:
        return "5 Years"
    else:
        return "More Than 5 Years"

df["tenure_group"]=df["tenure"].apply(lambda x: tenure(x))
```

```python
tenure_group_churn_yes=df[df["Churn"]=="Yes"]
tenure_group_churn_yes["tenure_group"].value_counts().plot.pie(autopct='%1.1f%%',wedgeprops = {"edgecolor":'black','linewidth': 1})
plt.title("Churn Percentages")
plt.show()
```



*   The Higher churn rate is in Less Than one Year ,There could be many reasons for that,one reason might be bad customer service



```python
tenure_group_churn_no=df[df["Churn"]=="No"]
tenure_group_churn_no["tenure_group"].value_counts().plot.pie(autopct='%1.1f%%',wedgeprops = {"edgecolor":'black','linewidth': 1})
plt.title("Non-Churn Percentages")
plt.show()
```



*   The Lower churn rate is in More Than Five Year **( The Customer Become Loyal )**
